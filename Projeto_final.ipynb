```python
def apaga_dado(df):
    lista=[]
    for i in range(len(df)):
        if (df.loc[i,'municipio']=='5' or df.loc[i,'municipio']=='3' or df.loc[i,'municipio']=='9' or df.loc[i,'municipio']=='8' or df.loc[i,'municipio']=='2' or df.loc[i,'municipio']=='1' or df.loc[i,'municipio']=='7' or df.loc[i,'municipio']=='4' or df.loc[i,'municipio']=='6'):
            lista.append(i)
            
    df=df.drop(lista,0)
    df['Indice'] = list(range(len(df)))
    
    df=df.set_index('Indice')
    
    return df
   
```
